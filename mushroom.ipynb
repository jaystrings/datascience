```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
#In this project. we will use a public dataset from kaggle.com, the “Mushroom Classification” dataset, and try to figure out if a mushroom is edible or not. 
#This dataset includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms. 
#Each species is identified as edible or poisonous. When it comes to mushrooms, there is no simple rule for determining the edibility.

mushroom_data = pd.read_csv('/Users/macbook/Downloads/mushrooms.csv')
mushroom_data.head()
```

```python
#i need a proper information about the mushroom data

mushroom_data.info()
```

```python
#Jeez all data in the dataset are alphabets, man i need to make them numerical. one way bro.. LabelEncoder()
#converting categorial data into numerical. Since all data in dataset is categorical, i can as well loop all to be numerical

labelencoder = LabelEncoder()
for i in mushroom_data.columns:
    mushroom_data[i] = labelencoder.fit_transform(mushroom_data[i])
mushroom_data

#Poisionous = 1
#Edible = 0
```

```python
#time to train and test our model.
x = mushroom_data.drop('class', axis=1)
y = mushroom_data['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(x_train, y_train)
```

```python
#predict for test data
y_pred = logistic_model.predict(x_test)
print(y_pred)
```

```python
#get the score
score_model = logistic_model.score(x_test, y_test)
print('\nScore :', score_model)
# Roughly 95% accuracy
```

```python
#check the confusion matrix

cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)
```
