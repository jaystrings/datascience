```python



from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
#creating the iris object called iris, with the iris data

iris =load_iris()

#Creating a dataframe with the four feature variables

df = pd.DataFrame(iris.data, columns= iris.feature_names)

#viewing the top 5 rows
df.head()
```

```python
#Adding a new colum for the species name
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
#viewing the top 5 rows
df.head()
```

```python
# Creating Test and Train Data
df['is_train'] = np.random.uniform(0, 1, len(df)) <= 0.75

# View the top 5 rows
df.head()
```

```python
#creating datframes with the test and trainning rows
train, test = df[df['is_train']==True], df[df['is_train']==False]

#Show the number of observation for the test and training dataframes

print('Number of observations in the trainnig data:', len(train))
print('Number of observations in the test data:', len(test))
```

```python
#create a list of feature columns names:
features = df.columns[:4]
#view features
features
```

```python
#converting  each species name into digits

y = pd.factorize(train['species'])[0]
y
```

```python

#creating a random forest classifier

clf = RandomForestClassifier(n_jobs = 2, random_state = 0)

#Training the classifier
clf.fit(train[features], y)
```

```python
#Applying the trained Classifier to the test.

test[features]
```

```python
#Applying the trained classifier to the test.
clf.predict(test[features])
```

```python
#viewing the predicted probablilities of the first 10 onservations.

clf.predict_proba(test[features][0:10])
```

```python
#mapping names for the plants for each predicted plant class
preds = iris.target_names[clf.predict(test[features])]

#View the predicted species for the first five observations,

preds[0:5]
```

```python
test['species'].head()
```

```python
#Creating the confusion matrix.

pd.crosstab(test['species'], preds, rownames = ['Actual Species'], colnames = ['Predicted Species'])
```
