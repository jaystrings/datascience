```python

import numpy as np
import pandas as pd
import pandas_profiling
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

from subprocess import check_output
```

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Casestudy/titanic_train.csv')
titanic_data.head()
```

```python
#Preprocessing the data.
#Dropping/Replacing missing entries of Embarked
#Replacing missing Values of Age and Fare with median values.
#Dropping the column 'Cabin' as it has too many null values
```

```python
titanic_data.info()
```

```python
891-titanic_data.count() #to check for the missing value.
```

```python
#Age has 177 missing values, it wwill be reasonable to replace this with the median, as it helps keep my data balanced and avoids distorting the regression or classification model.
# Cabin will be dropped because When a feature has more than ~50% missing, it’s usually not worth trying to impute because:
#There’s too little real information to make good guesses.
#Embarked has 2 missing value, and mode may just be the right choice to use in this case, why ? 
'''The mode is simply the most common category in that column.

Since 'S' appears most often, it’s statistically the most likely port those 2 missing passengers came from.

It preserves the data pattern and doesn’t distort category proportions.'''


```

```python
#start with the Embarked. 

titanic_data.Embarked = titanic_data.Embarked.fillna(titanic_data['Embarked'].mode()[0], inplace = True)
```

```python
#Dropping cabin

titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
#cleaning missing values in Age. 
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.info()
```

```python
#we csn see that all values have been added and data can be considered clean. 
 
```

```python
#Creating a family size, remember we have columns with parent and children, siblings and spouse. 

titanic_data['FamilySize']= titanic_data['SibSp'] + titanic_data['Parch']+1

#I added plus one because for one to die, it means at somepoint he was alive
```

```python
titanic_data.head()
```

```python
#Segmenting a column as per age, Age less than 15 as child, Age greater than 15 as Males and Females as per thier gender

titanic_data['GenderClass'] = titanic_data.apply(lambda x: 'child' if x ['Age']< 15 else x['Sex'], axis =1)
```

```python
titanic_data.head()
```

```python
titanic_data[titanic_data.Age<15].head(2)
```

```python
#Dummification of GenderClass & Embarked
#Why do we need to do dummification? Dummification also known as oneHot Encoding means converting categorical variables (like Sex, Embarked, GenderClass, etc.) into numerical form (0s and 1s) so that machine learning algorithms can understand them.

```

```python
titanic_data.head()
```

```python
titanic_data = pd.get_dummies(titanic_data, columns = ['GenderClass', 'Embarked'], drop_first = True)
```

```python
#So i will be dropping some columns because they are unique,it doesn’t help predict survival. Encoding would create thousands of useless dummy columns.
#Name, sex, Age, sibsp, parch, ticket. 
```

```python
titanic = titanic_data.drop(['Name', 'Ticket','Sex', 'SibSp', 'Parch', 'PassengerId'], axis = 1)
titanic.head()
```

```python
#one issue here, i still have my Gender class as boolean. so lets convert this to numbers

# Convert all boolean columns to int
titanic = titanic.astype({col: int for col in titanic_data.select_dtypes('bool').columns})
```

```python
titanic.head()
```

```python
# Drawing pair plot to know the joint relationship between 'Fare', 'Age', 'Pclass' and 'Survived'

import seaborn as sns
import matplotlib.pyplot as plt

# Set a dark background and style
sns.set(style="darkgrid")

# Define custom colors: red for not survived (0), black for survived (1)
custom_palette = {0: "red", 1: "black"}

# Draw the pairplot
g = sns.pairplot(
    titanic_data[['Fare', 'Age', 'Pclass', 'Survived']],
    hue='Survived',
    diag_kind='hist',
    dropna=True,
    palette=custom_palette,
    plot_kws={'alpha': 0.7, 's': 45},
    height=3,         # adjust individual plot size
    aspect=1.1        # width-to-height ratio
)

# Add more spacing between plots
g.fig.subplots_adjust(wspace=0.25, hspace=0.25)

# Add title
plt.suptitle("Pairplot: Fare, Age, Pclass vs Survival", fontsize=14, y=1.03)

plt.show()

```

```python
#To understand this pairplot better, i will proceed with more visualizations, starting with age:
#Age because, we need to know the age group, that had the highest survival rate.

# Create age bins
bins = [0, 12, 18, 30, 45, 60, 80]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Mid Age', 'Senior']

titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=bins, labels=labels)
```

```python
titanic_data.head()
```

```python
#Calaculate the Survival rate per group.

age_survival = titanic_data.groupby('AgeGroup')['Survived'].mean().reset_index()
age_survival
```

```python
#visuals
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.barplot(
    data=age_survival,
    x='AgeGroup',
    y='Survived',
    palette='coolwarm',
    edgecolor='black'
)

plt.title('Survival Rate by Age Group', fontsize=14)
plt.ylabel('Survival Rate')
plt.xlabel('Age Group')
plt.ylim(0,1)
plt.show()
```

```python
#Adding labels

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

titanic_data['AgeGroup'] = pd.cut(
    titanic_data['Age'],
    bins=[0, 12, 18, 30, 50, 80],
    labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
)

# Group and calculate survival rate
age_survival = titanic_data.groupby('AgeGroup')['Survived'].mean().reset_index()

# Plot
plt.figure(figsize=(8, 5))
bar = sns.barplot(x='AgeGroup', y='Survived', data=age_survival, palette='dark:b_r')

# Add title and axis labels
plt.title('Survival Rate by Age Group', fontsize=16, fontweight='bold')
plt.ylabel('Survival Rate')
plt.xlabel('Age Group')

# Add numeric labels on top of each bar
for i in range(len(age_survival)):
    plt.text(
        i, 
        age_survival['Survived'][i] + 0.02,  # little offset above the bar
        f"{age_survival['Survived'][i]*100:.1f}%",  # convert to percentage
        ha='center', va='bottom', fontsize=10, fontweight='bold', color='black'
    )

# Make it look clean
sns.despine()
plt.tight_layout()
plt.show()
```

```python
#highest survival rate are children, while lowest survival rates are the adults.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

# Compute survival rate by passenger class
titanic_pclass = titanic_data.groupby('Pclass', as_index=False)['Survived'].mean()

# Set the style
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))

# Barplot for survival rate by passenger class
ax = sns.barplot(
    data=titanic_pclass,
    x='Pclass',
    y='Survived',
    palette='coolwarm',
    edgecolor='black'
)

# Title and labels
plt.title('Survival Rate by Passenger Class', fontsize=14)
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.ylim(0, 1)

# Add numeric labels on each bar
for i, row in titanic_pclass.iterrows():
    plt.text(
        i,
        row['Survived'] + 0.02,
        f"{row['Survived']*100:.1f}%",
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold',
        color='black'
    )

# Clean up look
sns.despine()
plt.tight_layout()
plt.show()
```
